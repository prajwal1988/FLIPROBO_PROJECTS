```python
# importing all the required libraries

import selenium 
import pandas as pd
from selenium import webdriver

import time
from selenium.common.exceptions import NoSuchElementException 
```

```python
# connecting to the web driver
driver = webdriver.Chrome(r"//Users/prajwalkunder/Downloads/chromedriver") 
```

```python
# getting the webpage https://www.naukri.com/
url= "https://www.naukri.com/"
driver.get(url) 
```

```python
# searching the web_element
search_job = driver.find_element_by_id("qsb-keyword-sugg")
search_location = driver.find_element_by_id("qsb-location-sugg") 
```

```python
# Enter “Data Analyst” in “Skill,Designations,Companies” field and enter “Bangalore” in “enter the location” field.

search_job.send_keys("Data Analyst") 
search_location.send_keys("Bangalore")
```

```python
# Then click the search button.
search_btn = driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()
```

```python
# Then scrape the data for the first 10 jobs results you get.
titles_tag = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']") 
companies_tag = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
experience_tag = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
location_tag = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']") 
```

```python
# writing for loop for the scraped data to convert into text.

job_titles=[]
for i in titles_tag:
    job_titles.append(i.text)
    
company_name=[]
for i in companies_tag:
    company_name.append(i.text)

experience_list=[]
for i in experience_tag:
    experience_list.append(i.text)
    
location_list=[]
for i in location_tag:
    location_list.append(i.text) 
```

```python
# checking collected data length

print(len(job_titles),len(company_name),len(experience_list),len(location_list)) 
```

```python
# creating dataframe for the scraped data.

jobs = pd.DataFrame({}) 
jobs['title']=job_titles
jobs['company']=company_name
jobs['experience']=experience_list
jobs['location']=location_list
```

```python
# getting only 10 data by using range method.
df = pd.DataFrame(jobs)
company_jobs = df.drop(range(10,20)) 
```

```python
len(urls) 
```

```python
# Extracting job description data.
description_tag=[]
for i in urls:
    driver.get(i)
    time.sleep(1) 
    try:
        description=driver.find_element_by_xpath('//section[@class="job-desc"]')
        description_tag.append(description.text)
    except NoSuchElementException:
        description_tag.append('-') 
```
