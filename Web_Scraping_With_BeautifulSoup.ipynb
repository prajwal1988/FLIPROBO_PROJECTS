# 3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
!pip install bs4
!pip install requests

from bs4 import BeautifulSoup
import requests

page = requests.get('https://www.imdb.com/india/top-rated-indian-movies')
soup = BeautifulSoup(page.content) 

movie = soup.find_all("td", class_= "titleColumn") 
year = soup.find_all("span", class_="secondaryInfo")
rating = soup.find_all("td", class_="ratingColumn imdbRating") 
```

```python
print(movie,year,rating)
```

```python
IMDB_movie=[]
for i in movie:
    IMDB_movie.append(i.text) 

IMDB_year=[]
for i in year:
    IMDB_year.append(i.text) 

IMDB_rating=[]
for i in rating:
    IMDB_rating.append(i.text)
```

```python
IMDB_MOVIE=[] 
for i in IMDB_movie:
    IMDB_MOVIE.append(i.replace("\n","")) 
    
IMDB_YEAR=[]
for i in IMDB_year:
    IMDB_YEAR.append(i.replace("\n",""))
    
IMDB_RATING=[]
for i in IMDB_rating:
    IMDB_RATING.append(i.replace("\n",""))
```

```python
import pandas as pd
from pandas import DataFrame
```

```python
IMDB=pd.DataFrame({})
IMDB['MOVIE_NAME']=IMDB_MOVIE
IMDB["MOVIE_RATING"]=IMDB_RATING
IMDB["MOVIE_YEAR"]=IMDB_YEAR
```

```python
df = pd.DataFrame(IMDB)
IMDB_DATAFRAME = df.drop(range(100,250)) 
IMDB_DATAFRAME
```
